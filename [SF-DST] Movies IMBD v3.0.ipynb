# 36. У какого из режиссеров выше вероятность выпустить фильм в прибыли? (5 баллов)101
Варианты ответа:
1. Quentin Tarantino
2. Steven Soderbergh
3. Robert Rodriguez
4. Christopher Nolan
5. Clint Eastwood

```python
#считаем сколько всего каждый режисер снял фильмов
totaldir = data.groupby(['director'])['original_title'].count()
#делаем из сериз датафрейм
td = totaldir.reset_index()
#отфильтровываем исходный датафрейм только для тех фильмов, у ктр сборы выше бюджета, то есть с прибылью
profitfilms = data[data.profit > 0]
#считаем сколько всего каждый режисер снял прибыльных фильмов
profitdir = profitfilms.groupby(['director'])['original_title'].count()
# делаем из сериз дарафрейм 
pd = profitdir.reset_index()
#объединяем датафреймы
joined = td.merge(pd, on='director', how='left')
#меняем названия колонок
joined.columns = ['director','totalfilms','profitfilms']
#создаем новую колонку
joined['percentage'] = joined['profitfilms']/joined['totalfilms']
#отфильтровываем только с максимальным процентом прибыльных фильмов
joined[joined.percentage == joined.percentage.max()]
#ползаю по ответам и ищу процент прибыльных фильмов каждого режиссера
joined[joined.director == 'Quentin Tarantino']
answer_ls.append(1)
```
